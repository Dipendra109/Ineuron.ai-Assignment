```python
#1. Write a Python Program to Add Two Matrices?
print('1. Write a Python Program to Add Two Matrices?')
n1 = int(input('Enter rows of  first matrix')) #rows first matrix
n2 = int(input('Enter columns of first matrix')) #columns first matrix
m1 = int(input('Enter rows of second matrix'))   #rows second matrix
m2 = int(input('Enter columns of second matrix'))#columns second matrix
if n1 == m1 and n2==m2  :
    print('Right matrices dimensions for addition. Please enter matrices elements')  #informing user
    a_mtx=[]         # presuming a blank matrix to append.
    print("Enter the elements for first matrix : ") #asking user to enter elements
    for i in range(n1):  #running loop to append user input in matrix of defined dimension.
        #a matrix has column so each list item is itself a list. This list has elements whose idenx will act as column number. 
        row_el_a_mtx=[]                                      
        for j in range(n2): 
            #asking user to provide each element of matrix. size of this list will be equal to number of columns of matrix. 
            row_el_a_mtx.append(int(input('Enter first matrix elements')))           
        a_mtx.append(row_el_a_mtx)                      #appending each row to blank list defined as a_mtx.
    print('First Matrix is ' , '\n', a_mtx)
    #similarly second matrix is filled. 
    b_mtx=[]       
    print("Enter the elements for Second matrix : ")
    for i in range(m1): 
        row_el_a_mtx=[]                                      
        for j in range(m2): 
            row_el_a_mtx.append(int(input('Enter second matrix elements')))           
        b_mtx.append(row_el_a_mtx)                      
    print('Second Matrix is ' , '\n', b_mtx)
    
    add_mtx=[]  #dummy matrix for addition with all elements equl to zero. 
    for i in range(n1): 
        row_el_a_mtx=[]                                      
        for j in range(n2): 
            row_el_a_mtx.append(0)           
        add_mtx.append(row_el_a_mtx)                      
    print('Dummy Matrix for Addition is ' , '\n', add_mtx)
    # Commands for addiing two matrices.Only equl matrices from dimesnionality can be added.  
    for i in range(len(a_mtx)):
        for j in range(len(a_mtx[0])):
            add_mtx[i][j]= (a_mtx[i][j]+b_mtx[i][j])
    print('Addition of matrices is', '\n', add_mtx)
else :
    print('Wrong selection of matrix dimension.Restart program.')
```

```python
print('2. Write a Python Program to Multiply Two Matrices?') 
n1 = int(input('Enter rows of  first matrix')) #rows first matrix
n2=int(input('Enter columns of first matrix')) #columns first matrix
m1=int(input('Enter rows of second matrix'))   #rows second matrix 
m2=int(input('Enter columns of second matrix'))#columns second matrix 
if n2 == m1 :
    print('Right matrices dimensions for Multiplication. Please enter matrices elements') 
    #user input for first matrix. 
    a_mtx= []   
    print("Enter the elements for first matrix : ")
    for i in range(n1):
        row_el_a_mtx=[]
        for j in range(n2): 
            row_el_a_mtx.append(int(input('Enter first matrix elements')))           
        a_mtx.append(row_el_a_mtx)    
        #print(a_mtx)
    print('First Matrix is ' , '\n', a_mtx)
    #user input for second matrix
    b_mtx=[]       
    print("Enter the elements for first matrix : ")
    for i in range(m1): 
        row_el_a_mtx=[]                                      
        for j in range(m2): 
            row_el_a_mtx.append(int(input('Enter second matrix elements')))           
        b_mtx.append(row_el_a_mtx)                      
    print('Second Matrix is ' , '\n', b_mtx)
    
    mul_mtx=[]  #dummy matrix for multiplication
    for i in range(n1): 
        row_el_a_mtx=[]                                      
        for j in range(m2): 
            row_el_a_mtx.append(0)           
        mul_mtx.append(row_el_a_mtx)                      
    print('Dummy Matrix for Multiplication is ' , '\n', mul_mtx)
    #print(len(b_mtx[0]))
    #print(len(a_mtx))
    #applying multiplication thoerem for matrixes by denoting each elemet of new matrix by i, j , p where p is common dimension of two matrices. 
    for i in range(len(a_mtx)):
        for j in range(len(b_mtx[0])):
            for p in range(len(b_mtx)):
                mul_mtx[i][j] += (a_mtx[i][p]*b_mtx[p][j])
    print('Multiplication of matrices is', '\n', mul_mtx)
else :
    print('Wrong selection of matrix dimension.Restart program.')
```
